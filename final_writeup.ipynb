####Address clean-up
Only *one* address:street tag I could find needed cleaning up -- 'summit place  shopping centre' needed to be changed to 'Summit Place'.  It turns out that this was a place called "The Steaming Bean" cafe -- I verified their address online.  There were a few other odd addresses, but they were all "Highway 9".  A Highway 9 address is a valid address for places along the Highway just outside of Breckenridge.

I corrected The Steaming Bean's address after the data had been loaded into MongoDB:
```python
steamingbean = db.summitco.find_one({'address.street' : 'summit place  shopping  center'})pprint.pprint(steamingbean)
steamingbean['address']['street'] = 'Summit Place'
steamingbean['name'] = 'The Steaming Bean'
db.summitco.save(steamingbean)
```
 
####Mountain Peaks
A large majority of the mountain peaks in this map area show their elevation in meters with an xml tag: `<tag k="ele" v="3019"/>`. However, there were four peaks that did not show their elevations. Upon closer inspection, all of the affected peaks turned out to be duplicate nodes -- with the other half of each duplicate having complete elevation information.  So, I deleted these peaks from the database:
```python
db.summitco.remove({'natural':'peak', 'ele': None})
```

This leaves us with 82 peaks: 
```python
db.summitco.find({'natural':'peak'}).count()
```
```markdown
    82
```
 
####Chair lifts
Chair lifts are considered as aerialways: (`<tag k="aerialway" v="chair_lift"/>`). However, there are other types of aerialways in the data. There are also some sub-tags within aerialways, such as `aerialway:occupancy`. One aerialway had a typo `chari_lift` that needed to be fixed (fixed in the Python code writing to database).

I decided to make *all* aerialway tags as a subdocument, so the chair_lift tag became `'aerialway':{'type':'chair_lift'}. `
```python
pprint.pprint(db.summitco.find_one({'aerialway': {'type': 'chair_lift', 'occupancy' : '6' } }  ))
```
```markdown
    u'aerialway': {u'occupancy': u'6', u'type': u'chair_lift'},
    ...
    u'name': u'Colorado SuperChair',
```

One of the aerialways, Gondola One at Vail, was listed as a chair_lift, and needed to be changed to `type:gondola`:
```python
gondieOne = db.summitco.find_one({'name': 'Gondola One (1)'})
gondieOne['aerialway']['type'] = 'gondola'
db.summitco.save(gondieOne)
```

###Further Exploration

As I explored the data, I noted that, in particular, the map items related to recreation were very well represented.  The ski mountains in particular were well-documented, as were the mountain peaks, golf courses, bike trails, and the parking lots to use while playing at these amazing locations. However, other amenities were very poorly documented.  Most of the roads were from the [TIGER data](http://wiki.openstreetmap.org/wiki/TIGER), and many of the building outlines sourced Bing.  A minority of businesses could be found on the maps -- as I went down the streets of Breckenridge, it looked like approximately 1 in 5 of the restaurants I know about were actually on the map.  In fact, only 30 restaurants in the whole map area were listed!  These being large resort areas, this is a very small number.  It looks like people with recreation interests did most of the work on the map, leaving much to be desired with respect to other amenities.  Compare and contrast the level of detail noted between the ski runs' data, the peaks' data, and the ameneties' data in the queries below.

####Number of Downhill Pistes at each Difficulty Level:
```python
db.summitco.aggregate([ { '$match' : { 'piste.type' : 'downhill' } },
                                { '$group' : { '_id' : '$piste.difficulty', 'count': { '$sum' : 1 } } },
                                { '$sort' : { 'count' : -1 } }, { '$limit' : 4 }      
                                ])
```
```markdown
    {u'_id': u'advanced', u'count': 195}
    {u'_id': u'intermediate', u'count': 181}
    {u'_id': u'easy', u'count': 119}
    {u'_id': u'expert', u'count': 20}
```

This is for black, blue, green, and double-black diamond runs, respectively.  These seem like reasonable, and relatively complete, numbers for the five ski areas of concern.  The count of "expert" runs seems a bit low though compared to the others.  

####Top 5 Peaks' Elevation, in Feet and Meters:
```python
top_peaks = db.summitco.aggregate([ { '$match' : { 'natural' : 'peak', 'name' : { '$ne': None } } },
                                   { '$project' : { '_id' : '$name', 'ele_meters' : '$ele' ,
                                                   'ele_feet' : { '$multiply' : ['$ele', 3.28084 ] } } },
                                   { '$sort' : { 'ele_feet' : -1 } }, { '$limit' : 5 }
                                   ])
```
```markdown
    {u'_id': u'Pacific Peak', u'ele_feet': 13868.11068, u'ele_meters': 4227.0}
    {u'_id': u'Crystal Peak', u'ele_feet': 13822.17892, u'ele_meters': 4213.0}
    {u'_id': u'Bald Mountain', u'ele_feet': 13638.45188, u'ele_meters': 4157.0}
    {u'_id': u'Tenmile Range Peak 10', u'ele_feet': 13615.486, u'ele_meters': 4150.0}
    {u'_id': u'Father Dyer Peak', u'ele_feet': 13595.80096, u'ele_meters': 4144.0}
```

There are no "fourteeners" within this map view (peaks over 14,000 feet), but two peaks come within 200 feet of the distinction.

####Longest Chair Lifts by Duration; Chair Lifts Without Listed Duration
```python
db.summitco.aggregate([ { '$match' : { 'aerialway.type' : 'chair_lift' } } ,
                                  { '$project' : { '_id' : '$name', 'duration' : '$aerialway.duration',
                                                  'occupancy' : '$aerialway.occupancy' } },
                                  { '$sort' : {'duration': -1 }}, { '$limit': 5}                
                                ])

# Many lifts didn't list a duration.  How many?
print '\nChair Lift Duration Listed: ' + str(db.summitco.find({'aerialway.type':'chair_lift', 
                                                             'aerialway.duration' : {'$ne': None } } ).count())

print '\nChair Lift Duration Not Listed: ' + str(db.summitco.find({'aerialway.type':'chair_lift', 
                                                             'aerialway.duration' : {'$eq': None } } ).count())
```

###Conclusion
After fairly thoroughly exploring the data for the Summit County (plus Vail) ski resort areas and surrounding towns, it is apparent that recreation-lovers have done a good job mapping the key recreation facilities.  However, much of the rest of the map remains fairly incomplete -- very few businesses, restaurants, and other useful amenities are mapped.  Some finer detail could be added to a few of the ski lifts and runs (duration, occupancy, type of grooming), but for the average user, the data are fairly complete to help them get to their adventure destination. But then what does a person do for *aprés ski*?  Some of this mapping may be fairly easy to do for someone willing to walk around these small towns with a GPS unit, to add a number of nodes.  The buildings are well-mapped, so someone just needs to add the nodes for these other amenities.  Then, the map would be useful for the day-time fun, as well as for the fun once the sun goes down.
